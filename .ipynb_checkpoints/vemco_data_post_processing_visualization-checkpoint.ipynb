```python
# Read in the data
# 
# for read_csv, use header=0 when row 0 is a header row 

filename = '\\Long_Beach_6_29_2021\\data1.csv'
df = pd.read_csv(filename, header=0)   # read the file w/header row #0
print(f"{filename} : file read into a pandas dataframe.")
```

```python
df_clean = df.dropna()
```

```python
df_clean
```

```python
# Plot using Seaborn
sns.lmplot(x='Distance (m)', y='Signal Level (dB)', fit_reg=True, data=df_clean, hue='Transmitter ID Number')
 
# Tweak using Matplotlib
plt.ylim(30, None)
plt.xlim(0, 58)
plt.savefig("phake_lake_signal_plot_enlarged.png")
```

```python
# Plot using Seaborn
sns.lmplot(x='Distance (m)', y='Noise-Level (dB)', fit_reg = True, data=df_clean, hue='Transmitter ID Number')
 
# Tweak using Matplotlib
plt.ylim(18, None)
plt.xlim(0, 58)
plt.savefig('phake_lake_noise_plot_enlarged.png')
```

```python
# Plot using Seaborn
sns.lmplot(x='Distance', y='Defense', data=df_clean, hue='ID')
 
# Tweak using Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
# Plot using Seaborn
sns.lmplot(x='Distance', y='Defense', data=df_clean, hue='ID')
 
# Tweak using Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```
